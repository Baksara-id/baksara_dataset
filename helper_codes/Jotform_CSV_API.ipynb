```python
import os
import requests
import csv

def download_images_from_csv(csv_file):
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)
        column_paths = {}

        # Create folder for each column
        for header in headers:
            column_folder = os.path.join('image_data', header)
            os.makedirs(column_folder, exist_ok=True)
            column_paths[header] = column_folder

        # Download images for each row
        for row in reader:
            for i, header in enumerate(headers):
                image_url = row[i]
                column_folder = column_paths[header]

                # Extract filename from the URL
                filename = image_url.split('/')[-1]

                # Download the image and save it to the column folder
                response = requests.get(image_url)
                if response.status_code == 200:
                    image_path = os.path.join(column_folder, filename)
                    with open(image_path, 'wb') as image_file:
                        image_file.write(response.content)
                    print(f"Downloaded image {filename} for column {header}")
                else:
                    print(f"Failed to download image {filename} for column {header}")

    print("All images downloaded successfully!")

# Example usage
csv_file = 'dataset.csv'
download_images_from_csv(csv_file)
```

```python
import shutil

def zip_folder(folder_path, zip_path):
    shutil.make_archive(zip_path, 'zip', folder_path)

# Example usage
folder_path = 'image_data'
zip_path = 'image_data'
zip_folder(folder_path, zip_path)
```
